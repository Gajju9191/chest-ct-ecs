```python
import boto3
import zipfile
import os
from pathlib import Path

# Initialize S3 client
s3 = boto3.client('s3')

# Your bucket name (use the one you created)
BUCKET_NAME = 'chest-ct-raw-data'  # Replace with your actual bucket name
ZIP_FILE = 'chest-data.zip'  # Name of your zip file in S3
```

```python
# Download the zip file from S3
print(f"📥 Downloading {ZIP_FILE} from S3...")

try:
    s3.download_file(BUCKET_NAME, ZIP_FILE, ZIP_FILE)
    print(f"✅ Downloaded successfully: {ZIP_FILE}")
    
    # Check file size
    file_size = os.path.getsize(ZIP_FILE) / (1024 * 1024)  # MB
    print(f"📦 File size: {file_size:.2f} MB")
    
except Exception as e:
    print(f"❌ Download failed: {e}")
```
